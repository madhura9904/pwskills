```python
import seaborn as sns
```

```python
#1
Barplot-used to aggregate the categorical data according to some methods and by default its the mean. 
Boxplot-distribution of the quantitative data that represents the comparisons between variables.
Scatterplot-Distibution of data
Pairplot-Plotting of all attributes with respect to all attr
Histogram-Frequency of data with respect to any one attr
```

```python
#2
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
#3
titanic=sns.load_dataset('titanic')
sns.boxplot(x=titanic.pclass,y=titanic.age)
```

```python
sns.boxplot(x=titanic.pclass,y=titanic.fare)
```

```python
#4
diam=sns.load_dataset('diamonds')
sns.histplot(x=diam['price'],hue=diam.cut)
```

```python
#5
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue=iris.species)
```

```python
#6
flyt=sns.load_dataset('flights')
sns.heatmap(data=flyt,index=flyt.year, columns=flyt.month,values=flyt.passengers)
```
