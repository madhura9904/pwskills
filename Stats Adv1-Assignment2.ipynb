```python
'Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.'
import numpy as np
import scipy.stats as stats

# Input sample data
sample1 = [12, 14, 15, 13, 16, 18, 15]
sample2 = [10, 11, 9, 12, 10, 8, 11]

# Hypothesis Test Parameters
alpha = 0.05  # significance level
equal_variances = True  # Assume equal variances if True, else False

# Calculate sample statistics
n1, n2 = len(sample1), len(sample2)
mean1, mean2 = np.mean(sample1), np.mean(sample2)
std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

# Print sample statistics
print("Sample 1 Mean:", mean1, "Std Dev:", std1, "Size:", n1)
print("Sample 2 Mean:", mean2, "Std Dev:", std2, "Size:", n2)

# Perform the t-test
if equal_variances:
    # Pooled t-test
    pooled_variance = (((n1 - 1) * std1**2) + ((n2 - 1) * std2**2)) / (n1 + n2 - 2)
    se = np.sqrt(pooled_variance * (1 / n1 + 1 / n2))
    t_stat = (mean1 - mean2) / se
    df = n1 + n2 - 2
else:
    # Welch's t-test
    se = np.sqrt((std1**2 / n1) + (std2**2 / n2))
    t_stat = (mean1 - mean2) / se
    df = ((std1**2 / n1) + (std2**2 / n2))**2 / (
        ((std1**2 / n1)**2 / (n1 - 1)) + ((std2**2 / n2)**2 / (n2 - 1))
    )

# Calculate p-value
p_value = 2 * stats.t.sf(np.abs(t_stat), df)

# Print results
print("\nHypothesis Test Results:")
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df:.2f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"Reject the null hypothesis at α = {alpha}.")
    print("There is a significant difference between the two population means.")
else:
    print(f"Fail to reject the null hypothesis at α = {alpha}.")
    print("There is no significant difference between the two population means.")
```

```python
#10
import numpy as np
import scipy.stats as stats

# Define the two samples
sample1 = [12, 14, 15, 13, 16, 18, 15]
sample2 = [10, 11, 9, 12, 10, 8, 11]

# Verify the samples have equal size
if len(sample1) != len(sample2):
    raise ValueError("Samples must have the same size for this implementation.")

# Calculate sample statistics
n = len(sample1)
mean1, mean2 = np.mean(sample1), np.mean(sample2)
std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((n - 1) * std1**2 + (n - 1) * std2**2) / (2 * n - 2))

# Calculate the t-statistic
t_stat = (mean1 - mean2) / (pooled_std * np.sqrt(2 / n))

# Calculate degrees of freedom
df = 2 * n - 2

# Calculate the p-value (two-tailed)
p_value = 2 * stats.t.sf(np.abs(t_stat), df)

# Print results
print("Two-Sample T-Test Results:")
print(f"Sample 1 Mean: {mean1:.4f}, Std Dev: {std1:.4f}")
print(f"Sample 2 Mean: {mean2:.4f}, Std Dev: {std2:.4f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"Reject the null hypothesis at α = {alpha}.")
    print("There is a significant difference between the population means.")
else:
    print(f"Fail to reject the null hypothesis at α = {alpha}.")
    print("There is no significant difference between the population means.")
```

```python
#16
import numpy as np
from scipy.stats import t

# Given data
mean1, mean2 = 80, 75
std1, std2 = 10, 8
n1, n2 = 30, 40
alpha = 0.01

# Calculate the test statistic
t_statistic = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))

# Degrees of freedom (using the Welch-Satterthwaite equation)
df = ((std1**2 / n1 + std2**2 / n2) ** 2) / (((std1**2 / n1) ** 2) / (n1 - 1) + ((std2**2 / n2) ** 2) / (n2 - 1))

# Calculate the critical t-value (two-tailed test)
t_critical = t.ppf(1 - alpha / 2, df)

# Calculate the p-value
p_value = 2 * t.sf(np.abs(t_statistic), df)

# Print the results
t_statistic, t_critical, p_value, p_value < alpha

```
