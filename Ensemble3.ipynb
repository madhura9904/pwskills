```python

---

### **Q1. How does bagging reduce overfitting in decision trees?**

Decision trees are **high-variance models**â€”they easily overfit to training data.  
**Bagging** (Bootstrap Aggregating) reduces overfitting by:

- **Training multiple decision trees** on different bootstrapped samples of the data.
- **Averaging (regression)** or **voting (classification)** their predictions.
- This **cancels out the noise** from any single tree, smoothing the prediction.



---

### **Q2. What are the advantages and disadvantages of using different types of base learners in bagging?**

| **Base Learner**     | **Advantages**                                           | **Disadvantages**                                      |
|----------------------|----------------------------------------------------------|--------------------------------------------------------|
| **Decision Trees**    | High variance â†’ works great with bagging                | Can be slow to train if the dataset is large           |
| **KNN**              | Can benefit from reduced variance via bagging            | Computationally expensive with large ensembles         |
| **Linear Models**    | Fast, easy to interpret                                  | Low variance â†’ bagging has minimal effect              |
| **SVMs**             | Powerful classifiers                                     | Training time + bagging adds complexity                |


---

### **Q3. How does the choice of base learner affect the bias-variance tradeoff in bagging?**

- Bagging mainly **reduces variance**.
- So it works **best with high-variance, low-bias models** like decision trees.
- If your base learner has **high bias** (e.g., a linear model on non-linear data), bagging wonâ€™t help much.

ðŸ“Œ **Ideal learner for bagging** = high-variance, low-bias.

---

### **Q4. Can bagging be used for both classification and regression tasks? How does it differ in each case?**


| **Task**         | **Aggregation Method**                  |
|------------------|------------------------------------------|
| **Classification** | **Majority vote** among base models       |
| **Regression**      | **Average** of base model predictions     |

The training process is the sameâ€”just the way predictions are combined differs.

---

### **Q5. What is the role of ensemble size in bagging? How many models should be included in the ensemble?**

- A larger ensemble generally improves performance by reducing variance.
- But beyond a certain point (usually **100â€“200 models**), performance **plateaus**.
- Too many models â†’ higher **computational cost** and **memory usage**.


---

### **Q6. Example of a real-world application of bagging in machine learning**


- **Random Forests** (a bagging method) are widely used for spam filtering.
- Each tree predicts spam/not-spam, and the final decision is based on **majority vote**.
- Bagging reduces the risk of overfitting noisy word frequency features.

- **Loan default prediction**
- **Medical diagnosis** (e.g., predicting disease presence)
- **Customer churn prediction**
- **Sentiment analysis**

---

```
