```python
#Matplotlib is a library that helps in data visualisation by plotting of the data in various graphical forms.
#Barplot,scatterplot,boxplot,lineplot,pie chart.
```

```python
#Scatter plots are type of graphs that shows relation between two variables
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
#Subplot function is used to plot different graphs on the same plot.
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=2,ncols=2)
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
axes[0][0].plot(x,y)
x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
axes[0][1].plot(x,y)
x= np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
axes[1][0].plot(x,y)
x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
axes[1][1].plot(x,y)
plt.show()
```

```python
#Barplot is used to plot numerical data where x and y represent to different variables and also to know the relationship between them
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.barh(company,profit)
plt.show()
```

```python
#Boxplot is used to plot categorical data.It shows the distribution of the quantitative data that represents the comparisons between variables. boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution i.e. the dots indicating the presence of outliers.
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data,notch=None,vert=None,patch_artist=None,widths=None)
plt.show()
```
